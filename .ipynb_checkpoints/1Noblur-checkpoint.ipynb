## 不进行模糊操作，直接二值化

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
# 读取彩色图像
img_bgr = cv.imread('./data/test2.jpg')
show(img_bgr)
```

```python
# 读取灰度图像
img_gray = cv.imread('data/test2.jpg', 0)
show(img_gray)
```

```python
plt.hist(img_gray.ravel(), bins=np.arange(-0.5, 256, 1))
plt.show()
```

```python
th, img_bin = cv.threshold(img_gray, -1, 255, cv.THRESH_TRIANGLE)
th, img_bin = cv.threshold(img_bin, th, 255, cv.THRESH_BINARY_INV)
print(th)
show(img_bin)    
```
