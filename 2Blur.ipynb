## 均值模糊、高斯模糊、中值滤波

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()

def imshow(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
# 读取灰度图像
img_gray = cv.imread('data/test2.jpg', 0)
show(img_gray)
```

### 均值模糊

```python
# 均值滤波
img_mean3x3 = cv.blur(img_gray, ksize=(3, 3))
img_mean5x5 = cv.blur(img_gray, ksize=(5, 5))
img_mean7x7 = cv.blur(img_gray, ksize=(7, 7))

# 保存图片
plt.subplot(131), imshow(img_mean3x3), plt.title('3x3')
plt.subplot(132), imshow(img_mean5x5), plt.title('5x5')
plt.subplot(133), imshow(img_mean7x7), plt.title('7x7')
plt.suptitle('Mean Blur')
plt.savefig('./data/Mean Blur.jpg')
```

```python
# 三角法阈值分割
th3x3, img_bin3x3 = cv.threshold(img_mean3x3, -1, 255, cv.THRESH_TRIANGLE)
th3x3, img_bin3x3 = cv.threshold(img_bin3x3, th3x3, 255, cv.THRESH_BINARY_INV)
th5x5, img_bin5x5 = cv.threshold(img_mean5x5, -1, 255, cv.THRESH_TRIANGLE)
th5x5, img_bin5x5 = cv.threshold(img_bin5x5, th5x5, 255, cv.THRESH_BINARY_INV)
th7x7, img_bin7x7 = cv.threshold(img_mean7x7, -1, 255, cv.THRESH_TRIANGLE)
th7x7, img_bin7x7 = cv.threshold(img_bin7x7, th7x7, 255, cv.THRESH_BINARY_INV)
print(f"thresh3x3 = {th3x3}", f"thresh5x5 = {th5x5}", f"thresh7x7 = {th7x7}")

# 保存图片
plt.subplot(131), imshow(img_bin3x3), plt.title('3x3')
plt.subplot(132), imshow(img_bin5x5), plt.title('5x5')
plt.subplot(133), imshow(img_bin7x7), plt.title('7x7')
plt.suptitle('Mean Blur Bin')
plt.savefig('./data/Mean Blur Bin.jpg')
```

### 高斯模糊

```python
# 高斯模糊
sigma = 1
img_Gauss3x3 = cv.GaussianBlur(img_gray, (3, 3), sigma)
img_Gauss5x5 = cv.GaussianBlur(img_gray, (5, 5), sigma)
img_Gauss7x7 = cv.GaussianBlur(img_gray, (7, 7), sigma)

# 保存图片
plt.subplot(131), imshow(img_Gauss3x3), plt.title('3x3')
plt.subplot(132), imshow(img_Gauss5x5), plt.title('5x5')
plt.subplot(133), imshow(img_Gauss7x7), plt.title('7x7')
plt.suptitle(f"GaussianBlur (sigma = {sigma})")
plt.savefig(f"./data/GaussianBlur (sigma = {sigma}).jpg")
```

```python
# 三角法阈值分割
th3x3, img_bin3x3 = cv.threshold(img_Gauss3x3, -1, 255, cv.THRESH_TRIANGLE)
th3x3, img_bin3x3 = cv.threshold(img_bin3x3, th3x3, 255, cv.THRESH_BINARY_INV)
th5x5, img_bin5x5 = cv.threshold(img_Gauss5x5, -1, 255, cv.THRESH_TRIANGLE)
th5x5, img_bin5x5 = cv.threshold(img_bin5x5, th5x5, 255, cv.THRESH_BINARY_INV)
th7x7, img_bin7x7 = cv.threshold(img_Gauss7x7, -1, 255, cv.THRESH_TRIANGLE)
th7x7, img_bin7x7 = cv.threshold(img_bin7x7, th7x7, 255, cv.THRESH_BINARY_INV)
print(f"thresh3x3 = {th3x3}", f"thresh5x5 = {th5x5}", f"thresh7x7 = {th7x7}")

# 保存图片
plt.subplot(131), imshow(img_bin3x3), plt.title('3x3')
plt.subplot(132), imshow(img_bin5x5), plt.title('5x5')
plt.subplot(133), imshow(img_bin7x7), plt.title('7x7')
plt.suptitle(f"GaussianBlur Bin (sigma = {sigma})")
plt.savefig(f"./data/GaussianBlur Bin (sigma = {sigma}).jpg")
```

### 中值滤波

```python
# 中值滤波
img_median3 = cv.medianBlur(img_gray, ksize=3)
img_median5 = cv.medianBlur(img_gray, ksize=5)
img_median7 = cv.medianBlur(img_gray, ksize=7)

# 保存图片
plt.subplot(131), imshow(img_median3), plt.title('ksize = 3')
plt.subplot(132), imshow(img_median5), plt.title('ksize = 5')
plt.subplot(133), imshow(img_median7), plt.title('ksize = 7')
plt.suptitle('Median Blur')
plt.savefig('./data/Median Blur.jpg')
```

```python
# 三角法阈值分割
th3, img_bin3 = cv.threshold(img_median3, -1, 255, cv.THRESH_TRIANGLE)
th3, img_bin3 = cv.threshold(img_bin3, th3, 255, cv.THRESH_BINARY_INV)
th5, img_bin5 = cv.threshold(img_median5, -1, 255, cv.THRESH_TRIANGLE)
th5, img_bin5 = cv.threshold(img_bin5, th5, 255, cv.THRESH_BINARY_INV)
th7, img_bin7 = cv.threshold(img_median7, -1, 255, cv.THRESH_TRIANGLE)
th7, img_bin7 = cv.threshold(img_bin7, th7, 255, cv.THRESH_BINARY_INV)
print(f"thresh3x3 = {th3}", f"thresh5x5 = {th5}", f"thresh7x7 = {th7}")

# 保存图片
plt.subplot(131), imshow(img_bin3), plt.title('ksize = 3')
plt.subplot(132), imshow(img_bin5), plt.title('ksize = 5')
plt.subplot(133), imshow(img_bin7), plt.title('ksize = 7')
plt.suptitle('Median Blur Bin')
plt.savefig('./data/Median Blur Bin.jpg')
```
